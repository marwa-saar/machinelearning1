```python
# Ensure all values are strings
df['Blood Pressure'] = df['Blood Pressure'].astype(str)

# Keep only rows with the '/' separator
df = df[df['Blood Pressure'].str.contains('/')]

# Split into systolic and diastolic
df[['BP_Systolic', 'BP_Diastolic']] = df['Blood Pressure'].str.split('/', expand=True).astype(float)

# Drop original column
df = df.drop('Blood Pressure', axis=1)

# Check result
print(df[['BP_Systolic', 'BP_Diastolic']].head())
```

```python
df['BP_Systolic'] = df['BP_Systolic'].astype(int)
df['BP_Diastolic'] = df['BP_Diastolic'].astype(int)
```

```python
df.columns = df.columns.str.strip()
df
```

```python
df = pd.get_dummies(df, columns=['Diet'], drop_first=True)
df
```

```python
df[['Diet_Healthy', 'Diet_Unhealthy']] = df[['Diet_Healthy', 'Diet_Unhealthy']].astype(int)
df
```
