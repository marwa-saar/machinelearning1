```python
import kagglehub
import os 
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("piyushborhade/knn-algorithms")

print("Path to dataset files:", path)
```

```python
file_path = os.path.join(path, 'iphone_purchase_records.csv')
df = pd.read_csv(file_path)
df
```

```python
df.info()
df.describe()
df.head()
```

```python
df['Gender'] = df['Gender'].map({'Male':1, 'Female':0}) #male 1, female 0
df
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
df
```

```python
#
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Load dataset
X, y = load_iris(return_X_y=True)

# Define KNN model
knn = KNeighborsClassifier()

# Define parameter grid (values of k to try)
param_grid = {'n_neighbors': range(1, 25)}

# Grid Search with cross-validation
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')

# Fit model
grid_search.fit(X, y)

# Best K
print("Best K:", grid_search.best_params_['n_neighbors'])
print("Best Accuracy:", grid_search.best_score_)
```
