```python
# -----------------------------
# Decision Tree ML for Diabetes
# -----------------------------

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt


# 2️⃣ Remove rows with 'Other' in gender (if present)
print(df['gender'].unique())

# 3️⃣ Separate features and target
X = df.drop(columns=["diabetes"])
y = df["diabetes"]  # 0=no diabetes, 1=diabetes

# 4️⃣ Identify categorical and numerical columns
cat_features = ["gender", "smoking_history"]
num_features = [col for col in X.columns if col not in cat_features]

# 5️⃣ Preprocess categorical variables with one‑hot encoding
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", categorical_transformer, cat_features)
    ],
    remainder="passthrough"
)

# 6️⃣ Build the Decision Tree model inside a pipeline
clf = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("classifier", DecisionTreeClassifier(random_state=42))
])

# 7️⃣ Train‑test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 8️⃣ Train the model
clf.fit(X_train, y_train)

# 9️⃣ Predict on test data
y_pred = clf.predict(X_test)

# 🔟 Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# 1️⃣1️⃣ Visualize the trained tree
# Extract the tree from the pipeline
tree_model = clf.named_steps["classifier"]

plt.figure(figsize=(20, 10))
plot_tree(
    tree_model,
    filled=True,
    feature_names=clf.named_steps["preprocess"].get_feature_names_out(),
    class_names=["No Diabetes", "Diabetes"]
)
plt.title("Decision Tree for Diabetes Prediction")
plt.show()
```
